```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('Adoptable_Pets.csv')
df.iloc[0:67,:]

#This dataset contains a list of adoptable cats and dogs from the 
#Montgomery County Animal Services and Adoption Center in Maryland. 
```

```python
x = df['Pet Age']
plt.hist(x, bins=10, color='orange')
plt.title('Ages of Adoptable Pets')
plt.xlabel('Age in years')
plt.ylabel('Frequency')
plt.show()

#This histogram illustrates that the majority of the pets up for adoption 
#are between the ages of 1 and 2. Since the pet's age data is skewed right,  
#the best measure of central tendency would be the median.
```

```python
x = df.loc[lambda df: df['Animal Type'] == 'CAT', 'Pet Age']
y = df.loc[lambda df: df['Animal Type'] == 'DOG', 'Pet Age']

plt.hist(x, bins=10, density=True, color='yellow')
plt.hist(y, bins=10, density=True, color='blue', alpha=0.5)
plt.title('Normed histogram of Ages of Adoptable Animals')
plt.xlabel('Ages in years')

plt.show()

#These histograms show the frequency of ages between the adoptable cats and dogs. 
#Visually, it is obvious that of the adoptable animals, there are far more younger
#dogs available than cats.  Cats, on the other hand, are generally older.  Both
#cats and dogs have the same age range of 1 to 13.  
```

```python
n_grps = 2
cat_gender = (16, 8) 
dog_gender = (18, 25)

index=np.arange(n_grps)
bar_width=0.35

plt.bar(index,cat_gender,bar_width, color='yellow', label='Cat')
plt.bar(index + bar_width, dog_gender, bar_width, color ='blue', label='Dog')

plt.title('Frequency of Animals by Gender')
plt.xticks(index + (bar_width)*0.5, ('Female','Male'))
plt.legend()

plt.tight_layout()
plt.show()

#This bar graph represents how many female and male animals are available for 
#adoption based on animal type.  Overall, it is clear that there are more dogs
#available for adoption.  There are more male dogs available, and there are more
#female cats ready for adoption.  
```

```python
labels = 'Small', 'Medium', 'Large', 'XLarge'
cat_sizes = [4, 18, 2, 0]
dog_sizes = [3, 24, 15, 1]
explode = (0.05, 0.05, 0.05, 0.05)

fig, (ax1, ax2) = plt.subplots(1,2)
ax1.pie(cat_sizes, labels=labels, autopct='%1.1f%%',startangle=90, explode=explode)
ax1.axis('equal')
ax1.set_title('Cats by Size')


ax2.pie(dog_sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode)
ax2.axis('equal')
ax2.set_title('Dogs by Size')

plt.show()

#These two pie charts show the frequency of available animals by size.  For both cats and dogs,
#there are more medium-sized animals than any other size.  However, dogs tend to trend larger 
#than cats.
```
